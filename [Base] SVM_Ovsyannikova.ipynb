````python
from matplotlib.colors import ListedColormap

from sklearn.svm import LinearSVC, SVC
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score, cross_val_predict, cross_validate, train_test_split
from sklearn.metrics import make_scorer, accuracy_score, roc_auc_score
from sklearn.datasets import make_blobs, make_circles, make_moons
````

````python
def make_moons_cls(size=1000, d=2):
    X, y = make_moons(n_samples=size, noise=0.15)
    if d > 2:
        X = np.concatenate((X, np.random.normal(size=(size, d-2))), axis=1)
    return X, y

def make_circles_cls():
    X, y = make_circles()

def generate_data_with_imb_classes(size1=100, size2=10):
    X = np.r_[(
        np.random.normal(loc=1.0, size=(size1, 2)),
        np.random.normal(loc=0.5, size=(size2, 2))
    )]

    y = np.ones(len(X))
    y[-size2:] = 0
    return X, y
````

````python
def plot_separating_surface(X, y, cls, view_support=False, title=''):
    x_min = min(X[:, 0]) - 0.1
    x_max = max(X[:, 0]) + 0.1
    y_min = min(X[:, 1]) - 0.1
    y_max = max(X[:, 1]) + 0.1
    h = 0.005
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])

    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))
    Z = cls.predict(np.c_[xx.ravel(), yy.ravel()])

    plt.figure(figsize=(10, 10))
    if title:
        plt.title(title)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', s=40, cmap=cm_bright)
    if view_support:
        plt.scatter(X[cls.support_, 0], X[cls.support_, 1],
                    c=y[cls.support_], edgecolors='k', s=150, cmap=cm_bright)
    Z = Z.reshape(xx.shape)
    plt.xticks(())
    plt.yticks(())
    plt.contourf(xx, yy, Z, cmap=cm, alpha=.3)
    plt.show()
````

````python
columns = [f'feature_{i}' for i in range(X_train.shape[1])]
df_train = pd.DataFrame(X_train, columns=columns)
df_train['target'] = y_train

correlation_with_target = df_train.corr()['target'].sort_values(ascending=False)

feature_correlations = df_train.drop(columns='target').corr()

print("Correlation between features:")
print(feature_correlations)
print()
print()
print("Correlation with target:")
print(correlation_with_target)
````

````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("svc", SVC())
])

param_grid = {
    "svc__C": [0.1, 1, 10],
    "svc__kernel": ["linear", "rbf", "poly"],
    "svc__degree": [2, 3, 4]
}


grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring="accuracy", verbose=2, n_jobs=-1)

grid_search.fit(X_train[:, [0, 1, 3, 4]], y_train)

results = grid_search.cv_results_
for i in range(len(results["mean_test_score"])):
    print(f"iteration {i+1}:")
    print(f"Params: {results['params'][i]}")
    print(f"mean accuracy: {results['mean_test_score'][i]:.4f}")
    print("-" * 50)

print("Best params:", grid_search.best_params_)
````

````python
# Your code here
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(random_state=42, max_iter=1000))
])

param_grid = {
    'logreg__penalty': ['l1', 'l2'],
    'logreg__solver': ['liblinear'],
    'logreg__C': [0.01, 0.1, 1, 10, 100]
}

grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

results = grid_search.cv_results_
for i in range(len(results["mean_test_score"])):
    print(f"iteration {i+1}:")
    print(f"Params: {results['params'][i]}")
    print(f"mean accuracy: {results['mean_test_score'][i]:.4f}")
    print("-" * 50)

print("best params:", grid_search.best_params_)
print("best accuracy:", grid_search.best_score_)
````
